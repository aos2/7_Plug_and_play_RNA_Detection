```python
rep1a = lr1.value_counts().sort_index().reset_index()
rep1a = rep1a.rename(columns={'index': 'Split site', 'Split Site': 'Unsorted Raw R1'})
rep1a = rep1a.set_index('Split site')
rep1a.index.name = None

rep2a = lr2.value_counts().sort_index().reset_index()
rep2a = rep2a.rename(columns={'index': 'Split site', 'Split Site': 'Unsorted Raw R2'})
rep2a = rep2a.set_index('Split site')
rep2a.index.name = None

rep3a = lr3.value_counts().sort_index().reset_index()
rep3a = rep3a.rename(columns={'index': 'Split site', 'Split Site': 'Unsorted Raw R3'})
rep3a = rep3a.set_index('Split site')
rep3a.index.name = None

rep4a = lonr1.value_counts().sort_index().reset_index()
rep4a = rep4a.rename(columns={'index': 'Split site', 'Split Site': 'Sorted Raw R1'})
rep4a = rep4a.set_index('Split site')
rep4a.index.name = None

rep5a = lonr2.value_counts().sort_index().reset_index()
rep5a = rep5a.rename(columns={'index': 'Split site', 'Split Site': 'Sorted Raw R2'})
rep5a = rep5a.set_index('Split site')
rep5a.index.name = None

rep6a = lonr3.value_counts().sort_index().reset_index()
rep6a = rep6a.rename(columns={'index': 'Split site', 'Split Site': 'Sorted Raw R3'})
rep6a = rep6a.set_index('Split site')
rep6a.index.name = None


```

```python
zeros = pd.DataFrame([0]*420, columns = ['ind'])

df_all = pd.concat([zeros, rep1a, rep2a, rep3a, rep4a, rep5a, rep6a], axis=1)
df_all = df_all.fillna(0)
df_all = df_all.drop(['ind'], axis=1) #Don't need the initialized column anymore
df_all.reset_index(inplace=True)#need to make the index into a column (indicates split site)
df_all = df_all.rename(columns={'index': 'Split_site'})
df_all.to_csv('reads.csv')
df_all

```
